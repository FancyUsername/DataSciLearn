```python
import random as rnd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

impacts = np.array(list(range(-20, 21))) * 5
df = pd.DataFrame(np.matrix([np.random.choice(impacts, 10) for i in range(0, 100)]), columns=["Col " + str(i) for i in range(0, 10)])

df.describe()
```

```python
sns.pairplot(df)
plt.show()
```
