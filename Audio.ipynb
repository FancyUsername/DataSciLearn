```python
import math
from functools import reduce
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(12, 5))
x = np.arange(-5 * math.pi, 5 * math.pi, 0.1)
y = np.sum(np.sin(x * i) for i in range(1, 100))
plt.plot(x, y)
```

```python
import pyaudio
import numpy as np

p = pyaudio.PyAudio()

volume = 0.5     # range [0.0, 1.0]
fs = 44100       # sampling rate, Hz, must be integer
duration = 3.0   # in seconds, may be float
f = 440.0        # sine frequency, Hz, may be float
x = 2*np.pi*np.arange(fs*duration)*f/fs

stream = p.open(format=pyaudio.paFloat32,
                channels=1,
                rate=fs,
                output=True)

for n in range(2, 6):
    samples = np.sum(np.sin(x * i) for i in range(1, n)).astype(np.float32)
    stream.write(volume*samples)

stream.stop_stream()
stream.close()
p.terminate()
```
